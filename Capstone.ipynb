<h3>Scraping wikipedia page

```python
import pandas as pd
 
df = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M",header=0,keep_default_na=False)
df1=df[0][df[0].Borough.str.contains('Not assigned') == False]
df1
```

<h3>Combining Neighbourhood values

```python

df2=df1.groupby('Postcode').agg(dict(Borough = 'min', Neighbourhood = lambda x: ', '.join(x)))

df2.reset_index()
```

<h3>Replacing Not assigned Neighbourhood value, Inserting rows in new Dataframe as changing existing Dataframe is not good practice

```python

for ind,row2 in df2.iterrows():
    if(row2['Neighbourhood'] == 'Not assigned'):
        row3['Neighbourhood'] = row2['Borough']
        row3['Postcode'] = row2['Postcode']
        row3['Borough'] = row2['Borough']
    else:
        row3=row2
    df3.append(row3)
        
print(df3)
```
